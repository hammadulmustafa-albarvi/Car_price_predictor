# Feature selection

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df_temp = df.copy()


for i in df_temp.select_dtypes(include=['object']).columns:
  oe = OrdinalEncoder()
  df_temp[i] = oe.fit_transform(df_temp[[i]])


```

```python
X = df_temp.drop('price',axis=1)
y = df_temp['price']
```

# Correlation

```python
sns.heatmap(df_temp.corr())
```

```python
f1 = df_temp.corr()['price'].to_frame().reset_index().rename(columns={'index':'feature','price':'imp_corr'}).sort_values(by='imp_corr',ascending=False).iloc[1:]
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100,random_state=42)
rf.fit(X,y)
```

```python
f2 = pd.DataFrame({
    'feature':X.columns,
    'imp_rf':rf.feature_importances_
}).sort_values(by='imp_rf',ascending=False)
```

# Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
gb.fit(X,y)
```

```python
f3 = pd.DataFrame({
    'feature':X.columns,
    'imp_gb':gb.feature_importances_
}).sort_values(by='imp_gb',ascending=False)
```

# Permutation Importance

```python
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
```

```python
rf_ = RandomForestRegressor(n_estimators=100,random_state=42)
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
rf_.fit(X_train,y_train)
perm = permutation_importance(rf_,X_test,y_test,n_repeats=30,random_state=42)
```

```python
f4 = pd.DataFrame({
    'feature':X.columns,
    'imp_perm':perm.importances_mean
}).sort_values(by='imp_perm',ascending=False)
```

# RFE

```python
from sklearn.feature_selection import  RFE
```

```python
estimator = RandomForestRegressor()
rfe = RFE(estimator,n_features_to_select=X.shape[1],step=1)
rfe.fit(X,y)
```

```python
f5 = pd.DataFrame({
    'feature':X.columns,
    'imp_rfe':rfe.estimator_.feature_importances_
}).sort_values(by='imp_rfe',ascending=False)
```

```python
final_df = f1.merge(f2, on='feature') \
             .merge(f3, on='feature') \
             .merge(f4, on='feature') \
             .merge(f5, on='feature')
```

```python
final_df
```

```python
final_df = final_df.set_index('feature')
```
