# Price vs Color

```python
color_families = [
    "White",
    "Black",
    "Silver",
    "Grey",
    "Blue",
    "Red",
    "Green",
    "Brown",
    "Beige",
    "Gold",
    "Orange",
    "Yellow",
    "Purple",
    "Pink",
    "Maroon",
    "Burgundy",
    "Teal",
    "Ivory",
    "Champagne",
    "Bronze",
    "Titanium",
    "Graphite",
    "Charcoal",
    "Other"
]
```

```python
def assign_color(x):
  a = x.split()
  for i in a:
    if i in color_families:
      return i
```

```python
df = df.dropna(subset=['color'])
```

```python
df['color'] = df['color'].apply(assign_color)
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x=df['color'],y=df['price'])
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(15,6))
sns.boxplot(x=df['color'],y=df['price'])
plt.xticks(rotation=90)
```
