```python
links_of_cars = []
name = []
car_year = []
price = []
kilometer_travelled = []
fuel_type = []
transmission = []

link = []
register_in = []
color = []
assembly = []
engine = []
body_type = []

interior = []
exterior = []
safety = []
comfort = []
```

```python
for page in range(1,5):
  links_of_cars = []
  webpage = requests.get(f'https://www.pakwheels.com/used-cars/family-cars/587667?page={page}',headers=headers).text
  soup  = BeautifulSoup(webpage,'lxml')
  no_cars_per_page = len(soup.find_all('div',class_=f'well search-list clearfix ad-container page-{page} pos-rel'))
  for i in range(no_cars_per_page):
    links_of_cars.append(f"https://www.pakwheels.com/{soup.find_all('a',class_ = 'car-name ad-detail-path')[i]['href']}")
  for i in range(len(links_of_cars)):
    link.append(links_of_cars[i])
    a = BeautifulSoup(requests.get(links_of_cars[i],headers=headers).text,'lxml')

    try:
      name.append(a.find('h1').text.strip())
    except:
      name.append(np.nan)

    try:
      price.append(a.find('strong',class_='generic-green').text.strip() )
    except:
      price.append(np.nan)


    details = a.find('tr').find_all('p')
    try:
      car_year.append(details[0].text.strip() )
    except:
      car_year.append(np.nan)

    try:
      kilometer_travelled.append(details[1].text.strip() )
    except:
      kilometer_travelled.append(np.nan)

    try:
      fuel_type.append(details[2].text.strip() )
    except:
      fuel_type.append(np.nan)

    try:
      transmission.append(details[3].text.strip() )
    except:
      transmission.append(np.nan)



    imp = a.find('ul',class_='list-unstyled ul-featured clearfix').find_all('li',class_=None)
    try:
      register_in.append(imp[0].text)
    except:
      register_in.append(np.nan)


    try:
      color.append(imp[1].text)
    except:
      color.append(np.nan)


    try:
      assembly.append(imp[2].text)
    except:
      assembly.append(np.nan)


    try:
      engine.append(imp[3].text)
    except:
      engine.append(np.nan)

    try:
      body_type.append(imp[4].text)
    except:
      body_type.append(np.nan)





    my_dict = {}
    l = a.find_all('h3')
    types = a.find_all('ul',class_='list-unstyled car-feature-list nomargin')
    check = ['Comfort & Convenience','Safety & Security','Interior','Exterior']
    start = 0
    end = 4
    if l[0].text.strip() == 'PakWheels Inspection Report':
      start = 1
      end = 5

    not_find = False
    for x,k in enumerate(range(start,end)):

      if l[k].text.strip() not in check:
        break

      all = types[x]
      my_dict[l[k].text.strip()] = [j.get_text(strip=True) for j in all.find_all('li')]


    for x,i in enumerate(my_dict):
      match i:
        case 'Interior':
          interior.append(my_dict[i])

        case 'Exterior':
          exterior.append(my_dict[i])


        case 'Safety & Security':
          safety.append(my_dict[i])


        case 'Comfort & Convenience':
          comfort.append(my_dict[i])

    for i in ['Comfort & Convenience','Safety & Security','Interior','Exterior']:
      if i not in list(my_dict.keys()):
        if i == 'Comfort & Convenience':
          comfort.append(np.nan)

        elif i=='Safety & Security':
          safety.append(np.nan)

        elif i=='Interior':
          interior.append(np.nan)

        else:
          exterior.append(np.nan)


  time.sleep(1)
  print(page)

```
