# Price the most important one

```python
df['price'].describe()
```

```python
df['price'].isnull().sum()
```

```python
sns.histplot(df['price'],kde=True,bins=50)
```

```python
df['price'].plot(kind='box')
```

```python
print(df['price'].skew())
print(df['price'].kurtosis())
```

```python
df['price'].quantile([0.01,0.10,0.95,0.99])
```

```python
q1 = df['price'].quantile(0.25)
q3 = df['price'].quantile(0.75)
iqr = q3 - q1
```

```python
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

```python
print(upper_bound)
print(lower_bound)
```

```python
outlier = df[(df['price'] > upper_bound) | (df['price'] < lower_bound)]
```

```python
outlier.shape
```

```python
outlier['price'].describe()
```

```python
bins = [0,1,5,10,20]
pd.cut(df['price'],bins=bins).value_counts().sort_index().plot(kind='bar')
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
sns.histplot(df['price'],bins=50,kde=True)
plt.subplot(1,2,2)
sns.histplot(np.log1p(df['price']),bins=50,kde=True)
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
sns.boxplot(df['price'])
plt.subplot(1,2,2)
sns.boxplot(np.log1p(df['price']))
plt.show()
```
