# Check the price values


```python
df = df[~(df['price']=='Call for price')]
```

```python
df.shape
```

```python
def convert_price_to_crore(price):
  if price[2] == 'lacs':
    return round(float(price[1])/100,2)
  elif price[2] == 'crore':
    return round(float(price[1]),2)
```

```python
df_cleaned = df
```

```python
df_cleaned['price'] = df['price'].str.split(' ').apply(convert_price_to_crore)
```

# Interior

```python
interiors = {'Infotainment System':7, 'Head Up Display (HUD)':9, 'Android Auto':5, 'Apple CarPlay':5, 'Front Speakers':3, 'Rear Speakers':4, 'Steering Switches':4, 'Paddle Shifters':6}
```

```python
temp_df = df
```

```python
def interior_score(x):
  if pd.isna(x):
    return 0
  score = 0
  for i in interiors.keys():
    if  i in x:
      score = score + interiors[i]
  return score
```

```python
df_cleaned['interior_scores'] = df_cleaned['interior'].apply(interior_score)
```

# Exterior

```python
exterior =     {'Alloy Wheels': 6,
    'Front Fog Lights': 3,
    'LED Headlights': 7,
    'DRLs': 4,
    'Sun Roof': 8,
    'Panoramic Sunroof': 10}
```

```python
def exterior_score(x):
  if pd.isna(x):
    return 0
  score = 0
  for i in exterior.keys():
    if  i in x:
      if i in ['Panoramic Sunroof','Sun Roof']:
        score = score + 0
      else:
        score = score + exterior[i]

  if 'Panoramic Sunroof' in x:
    return score + 10
  elif 'Sun Roof' in x:
    return score + 8
  else:
    return score
```

```python

df_cleaned['exterior_scores'] = df_cleaned['exterior'].apply(exterior_score)
```

# Safety

```python
safety = {
    'Air Bags': 10,
    'ABS': 8,
    'Traction Control': 7,
    'Tyre Pressure Monitoring System (TPMS)': 5,
    '360 Camera': 8,
    'Rear Camera': 6,
    'Front Camera': 4,
    'Parking Sensors': 5,
    'Immobilizer Key': 6,
    'Power Locks': 3
}
```

```python
def safety_score(x):
  if pd.isna(x):
    return 0
  score = 0
  for i in safety.keys():
    if  i in x:
      score = score + safety[i]
  return score
```

```python
df_cleaned['safety_scores'] = df_cleaned['safety'].apply(safety_score)
```

# Comfort

```python
comfort = {
    'Ventilated Seats': 10,
    'Heated Seats': 8,
    'Power Seats': 7,
    'Climate Control': 7,
    'Cruise Control': 6,
    'Push Start': 6,
    'Keyless Entry': 5,
    'Rear AC Vents': 5,
    'Power Mirrors': 4,
    'Power Steering': 2,
    'Air Conditioning': 2
}
```

```python
def comfort_score(x):
  if pd.isna(x):
    return 0
  score = 0
  for i in comfort.keys():
    if  i in x and i not in ['Air Conditioning','Climate Control']:
      score = score + comfort[i]

  if 'Climate Control' in x:
    return score + 7
  elif 'Air Conditioning' in x:
    return score + 2
  else:
    return score
```

```python
df_cleaned['comfort_scores']  = df_cleaned['comfort'].apply(comfort_score)
```
